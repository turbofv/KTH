3. 승차인원에 비해서 하차 인원의 비율이 높은 역 5개를 찾으세요

```python
# 짝수 : 승차, 홀수 : 하차
data.head()
승차 = data.iloc[:,::2]
하차 = data.iloc[:,1::2]
```

5. 유임 하차에 비해 무임 하차 비율이 가장 높은 역 5개를 찾으세요

```python
df = pd.read_excel('2019년 05월  교통카드 통계자료.xls', sheet_name=2,encoding='euc-kr',skiprows=0,index_col=3)
df.shape
df
```

```python
df.head()
subway = df.iloc[:,3:-1]
subway.head()
```

```python
subway = subway.replace(',','',regex=True)
subway = subway.astype('int')
```

```python
유임하차 = subway['유임하차']
무임하차 = subway['무임하차']
비율 = 무임하차/유임하차
비율
```

```python
y = 비율.sort_values(ascending=False)[:5]
```

```python
#plt.bar(y.index,y)
plt.title('유임 하차에 비해 무임 하차 비율이 가장 높은 역 5개',fontsize=30)

plt.pie(y, labels=y.index,autopct='%d')
```
