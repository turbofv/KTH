```python
from sklearn.model_selection import train_test_split # 함수
```

```python
# 데이터를 학습용 75% / 테스트용 25% 나눠 섞습니다.
# 실험과 증명과 재현을 위해서 random_state 를 사용
# x_train,x_test,y_train, y_test = train_test_split(iris_dataset.data, iris_dataset.target)
x_train,x_test,y_train, y_test = train_test_split(iris_dataset.data, iris_dataset.target,random_state=0)

iris_dataset.data.shape, iris_dataset.target.shape
```

```python
 x_train.shape, x_test.shape,y_train.shape,y_test.shape
```

```python
df= pd.DataFrame(x_train, columns=iris_dataset.feature_names)
df
```

```python
plt.rc('font', family='Malgun Gothic')
df.plot()
```

```python
pd.plotting.scatter_matrix(df,c=y_train, figsize=(10,10), marker='o',hist_kwds={'bins':20}, s=60, alpha=.8, cmap=mglearn.cm3)
```

```python
# k-최근접 이웃 모델

from sklearn.neighbors import KNeighborsClassifier # 대문자로 시작하면 클래스
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn
```

```python
knn.fit(x_train, y_train) # fit (모델을 훈련 데이터에 맞게 피팅시켜라)
```

```python
plt.title(iris_dataset.feature_names[0])
plt.scatter(np.arange(len(x_train)),x_train[:,0], c=y_train, cmap=mglearn.cm3)
```

```python
x_new = np.array([[5,2.9,1,0.2]]) #야생에서 꽃을 발견헀다!
x_new
x_new.shape
```

```python
prediction = knn.predict(x_new) # 클래스를 예측(분류)해라
print('예측', prediction)
print('예측한 타겟의 이름은 ', iris_dataset.target_names[prediction])
```

```python
y_pred = knn.predict(x_test)
print("테스트 세트에 대한 예측값 ")
print('예측', y_pred)
print('정답', y_test)
```

```python
print('정확도',knn.score(x_test, y_test))
```
