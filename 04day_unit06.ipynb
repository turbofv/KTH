```python
import pandas as pd
data = pd.read_csv('seoul.csv',encoding='euc-kr')
data=data.dropna() # 전처리 : 누락된 값이 있으면 그 행은 버린다.
data.head(100)
```

```python
month = []
for i in range(12):
    month.append([])
x =np.array(data)
x.shape, x[0]
```

```python
np.mean(x[:,-1])
```

```python
for i in range(len(x)):
    row = x[i]
    date =row[0]
    high = row[-1]
    mon = date.split('-')[1]
    
    month[int(mon)-1].append(high)
```

```python
plt.figure(figsize=(10,5))
m= plt.boxplot(month)
```
